## Filtering

```python
#Dilter mask that applies to all the rows that meet the criteria
df['state'] == 'AZ'

#Set filter
flt =(df['state'] == 'AZ')
```

Apply filter

```python
df[flt]
```

```python
#uding .loc - has an option to specify the columns
df.loc[flt,['state', 'population']]
```

```python
#filter by population
highpop = (df['population'] > 10000000)
df.loc[highpop]
```

```python
#Choose specific states
#
#Male a list of the states
sts = ['DC','CA','CO','NY','GA']
#make the filter
flt_sts = df['state'].isin(sts)
df.loc[flt_sts]
```

```python
#filter name that has letter A in it
stname = df['state'].str.contains('A')
df[stname]
```

## Updating the data frame

```python
#Making column names upper case
df.columns = [x.upper() for x in df.columns]
df.tail()
```

```python
#replCE SPACES WITH _
df.columns = df.columns.str.replace (' ', '_')
```

```python
#Making column names lower case
df.columns = [x.lower() for x in df.columns]
df.tail()
```

```python
#rename columns
df.rename(columns={'fips': 'id'}, inplace=True)
df.head()
```
