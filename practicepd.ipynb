```python
#Access particular columns(series)
df['state']
```

```python
#Access seberal columns - series
sr = df [['state', 'county', 'level','actuals.cases', 'actuals.deaths', 'actuals.positiveTests',
       'actuals.negativeTests']]
#Display filtered dataframe
sr.head()
```

####  **iloc** refers to integer location therefore uses unteger indexes.

```python
#Access rows of dataframe
df.iloc[19]
```

```python
#Access rows of filtered dataframe
sr.iloc[19]
```

```python
#Access multiple rows
sr.iloc[6:12]
```

```python
#Access multiple rows
sr.iloc[[3,5,8,14,32]]
```

```python
#Access specific rows and columns
sr.iloc[[3,5,8,14,32], [0,1,4]]
```

#### For *loc* indexes can be strings

```python
#Access multiple rows
df.loc[[3,5,8,14,32],['state','actuals.hospitalBeds.capacity','actuals.deaths']]
```

When slicing .loc the last index is inclusive.

```python
df.loc[6:12, 'actuals.cases': 'actuals.newCases']
```
